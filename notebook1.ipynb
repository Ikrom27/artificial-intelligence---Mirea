### Задание 4.3.1

```python
import numpy as np
import matplotlib.pyplot as plt

# Задаем массив случайных значений
np.random.seed(0)
x = np.random.rand(100)

# Рассчитываем среднее значение
mean = np.mean(x)
print("Mean value:", mean)

# Рассчитываем медианное значение
median = np.median(x)
print("Median value:", median)

# Сравниваем результаты
print("Mean and median are equal:", mean == median)

# Построение точечной диаграммы рассеивания
plt.scatter(range(100), x)
plt.show()
```

Из полученных результатов мы можем сделать вывод, что в данном случае среднее и медианное значение различаются, так как распределение данных не является симметричным. Точечная диаграмма рассеивания поможет визуально оценить распределение данных.

### Задание 4.3.2

```python
def superFunc(x):
    return (np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x**2)) /
            np.abs(1 - np.sin(x)**3) + np.log(np.abs(2 * x)))
```

```python
results = []

for i in range(1, 11):
    results.append(superFunc(i))

print(results)

half = results[:5]

plt.plot(results)
plt.title("linear")

plt.scatter(range(5), half)
plt.title("Scatter")
```

### Задание 4.3.5

```python
def calculator(x, y, operation):
    if operation == '+':
        return x + y
    elif operation == '-':
        return x - y
    elif operation == '*':
        return x * y
    elif operation == '/':
        return x / y
    elif operation == 'exp':
        import math
        return math.exp(x + y)
    elif operation == 'sin':
        import math
        return math.sin(x + y)
    elif operation == 'cos':
        import math
        return math.cos(x + y)
    elif operation == '^':
        return x ** y
    else:
        return "Invalid operator. Please use either '+', '-', '*', '/', 'exp', 'sin', 'cos', or '^'"

x = float(input("Enter first number: "))
y = float(input("Enter second number: "))
operation = input("Enter an operator (+, -, *, /, exp, sin, cos, ^): ")

print("Result:", calculator(x, y, operation))
```
